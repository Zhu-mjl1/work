### 1.2 前向传播模型

````python
g = sigmoid

def my_dense(a_in, W, b):
    """
    Computes dense layer
    Args:
      a_in (ndarray (n, )) : Data, 1 example 
      W    (ndarray (n,j)) : Weight matrix, n features per unit, j units
      b    (ndarray (j, )) : bias vector, j units  
    Returns
      a_out (ndarray (j,))  : j units|
    """
    units = W.shape[1]
    a_out = np.zeros(units)
    for j in range(units):               
        w = W[:,j]                                    
        z = np.dot(w, a_in) + b[j]         
        a_out[j] = g(z)               
    return a_out
````

> 实现`Tensorflow's model.predict ()`

````python
def my_sequential(x, W1, b1, W2, b2):
    a1 = my_dense(x,  W1, b1)
    a2 = my_dense(a1, W2, b2)
    return a2
````

### 2.2 激活函数
#### 2.2.1 激活函数简介
* 常用的激活函数有sigmoid函数、ReLU函数、线性函数和softmax函数等
* sigmoid函数可以用于估计概率
* 线性函数可以直接将输入作为输出
* ReLU函数可以使神经元的输出取非负值 
    * $g(z) = max(0,z)$
    * 常用于多分类任务
    * eg.图像分类（非二元问题）
#### 2.2.2 如何选择激活函数
* 对于输出层：二分类 - sigmoid;回归且所有输入非负 - ReLU;回归且输入有正有负 - 线性
* 对于隐藏层：一般选择ReLU，除非二分类，原因是更便于梯度下降
#### 2.2.3 为什么需要激活函数
* 需要激活函数且不能全使用线性函数的原因：
    * 复合后仍然是线性函数，仍然是一个线性回归

### 2.3 多分类
####  2.3.1 multiclass
* 多类别分类是指分类问题中可能有多个输出标签
* Softmax回归算法是逻辑回归算法的一种推广，可以用于解决多类别分类问题。
#### 2.3.2 softmax
* softmax原理：
    * 对于每个类别$j$，计算一个线性模型的输出值$z_j$。$z_j = w_j * x_j + b_j$
    * 对于每个类别$j$，计算一个概率值$a_j$。$a_j = \frac{e^z_j}{\sum_{k=1} ^{N} e^{z_k}}$
    * 根据概率值$a_j$来预测输入样本类别
    * 输出$\mathbf{a}$是长度为N的向量，因此对于softmax回归还可以写作：

\begin{array}
\mathbf{a}(x) =
\begin{bmatrix}
P(y = 1 | \mathbf{x}; \mathbf{w},b) \\
\vdots \\
P(y = N | \mathbf{x}; \mathbf{w},b)
\end{bmatrix}
=\frac{1}{ \sum_{k=1}^{N}{e^{z_k} }}
\begin{bmatrix}
e^{z_1} \\
\vdots \\
e^{z_{N}} 
\end{bmatrix}
\end{array}

    



#### 2.3.3 softmax函数的tensorflow实现与改进
* 由于softmax函数的损失函数需要计算e为底数的量，这个数值可能会变得很大，由此出现浮点数的精度下降，可能产生严重的舍入误差。
* 为了解决这个问题，通常将输出层的`activation`设置成`linear`，而在`model.compile`中作出调整。即：

```python
model.compile(loss = SparseCategoricalCrossEntropy(from_logits = True))
```

#### 2.3.4 区分multiclass && multilabel
 **multiclass :**
* 将样本分为多个互斥的类别
* 每个样本只能属于一个类别
* eg. 将图像分类为猫、狗、鸟等

 **multilabel :**
* 将样本分配给多个可能的标签
* 每个样本可以属于一个或多个标签
* eg. 对于一篇新闻文章，可以将其标记为政治、体育、娱乐等，也可以只标记为政治


````python
#Lab_C2_W2_SoftMax
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from IPython.display import display, Markdown, Latex
from sklearn.datasets import make_blobs
%matplotlib widget
from matplotlib.widgets import Slider
tf.autograph.set_verbosity(0)
````

````python
def my_softmax(z):
    ez = np.exp(z) #这里的z和ez都是矢量
    sm = ez/np.sum(ez)
    return sm
# make  dataset for example
centers = [[-5, 2], [-2, -2], [1, 2], [5, -2]]
X_train, y_train = make_blobs(n_samples=2000, centers=centers, cluster_std=1.0,random_state=30)
````

````python
#第一种实现方式，误差会大一些
model = Sequential([
    Dense(25, activation = 'relu'),
    Dense(15, activation = 'relu'),
    Dense(4 , activation = 'softmax')
])
model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(0.001)
)
model.fit(
    X_train,y_train,
    epochs = 10
)
````

````python
p_1 = model.predict(X_train)
print("two example:\n",p_1[:2])
print("largest:",np.max(p_1),"smallest:",np.min(p_1))
````

````python
#第二种实现方式，更推荐
preferred_model = Sequential ([
    Dense(25, activation='relu'),
    Dense(15, activation='relu'),
    Dense(4 , activation='linear')
])

preferred_model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = tf.keras.optimizers.Adam(0.001),
)

preferred_model.fit(
    X_train,y_train,
    epochs = 10
)
````

### 2.4 Additional Neural Network Conceptions
#### 2.4.1 Adam优化器
* Adam: Adaptive Moment estimation
* 可以自动调整学习率，根据模型参数选择自行增大或者减小学习率
* 相比梯度下降，Adam更鲁棒
#### 2.4.2 Additional Layer Types
* 神经网络中的传统层类型是全连接层（dense layer），每个神经元都接收前一层的所有激活值作为输入
* 卷积层
    * 每个神经元只能看到输入的一部分区域
    * 优势:
        * 加速计算和减少训练数据需求
        * 减少过拟合的风险
* 卷积神经网络
    * 包含多个卷积层的神经网络，常用于图像和时间序列等领域的任务
    * 架构可以根据需求进行调整，如窗口大小和层中神经元的数量等
